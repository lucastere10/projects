```python
import matplotlib.pyplot as plt
#somatorio  
def sum(n, s = 0, i = 1):
    for i in range(1, n+1): s = s + 1/i
    return s
  
#multiplos n
n_terms = []
for n in range(1,11,1): 
    n_terms.append(round(sum(n), 6))
    print(f'With n = {n}, sum = ', round(sum(n), 6))

#graph
plt.plot(n_terms)

```
